```python
import pandas as pd
df.dropna(axis=1, how='all', inplace=True)

df.fillna({
    'driver_age' : df['driver_age'].median(),
    'search_type' : 'None',
    'stop_duration' : 'Unknown',
    'violation': 'Unknown',
    'stop_outcome' : 'Unknown',
}, inplace = True)

df['timestamp'] = pd.to_datetime(df['stop_date'] + ' ' + df['stop_time'])
```

```python
df
```

```python
import mysql

connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="12345678"
)

cursor = connection.cursor()
```

```python
cursor.execute("create database police")
```

```python
cursor.execute('use police')
```

```python

cursor.execute('''
    ALTER TABLE vehicles
    DROP COLUMN post_location''')
```

```python
cursor.execute('''
    CREATE TABLE police_logs (
    id INT AUTO_INCREMENT PRIMARY KEY,
    stop_date DATE,
    stop_time TIME,
    county_name VARCHAR(100),
    driver_gender VARCHAR(10),
    driver_age INT,
    driver_race VARCHAR(50),
    search_conducted BOOLEAN,
    search_type VARCHAR(100),
    drugs_related_stop BOOLEAN,
    stop_duration VARCHAR(50),
    vehicle_number VARCHAR(20),
    stop_outcome VARCHAR(100),
    violation VARCHAR(100),
    timestamp DATETIME
)
''')
```

```python
# Vehicles
cursor.execute('''create table vehicles (
    vehicle_number VARCHAR(20),
    owner_id INT,
    type VARCHAR(100),
    flagged_status VARCHAR(50)
    )
''')

print('Table created successfully!!')
```

```python
# Check Post Logs
cursor.execute('''
    CREATE TABLE check_post_logs(
        id INT,
        vehicle_number VARCHAR(20),
        driver_id INT,
        officer_id INT,
        timestamp TIMESTAMP,
        status VARCHAR(50)
    )
''')

print('Table created successfully!!')
```

```python
# Officers
cursor.execute('''CREATE TABLE officers (
    officer_id INT,
    name VARCHAR(100),
    post_location VARCHAR(100),
    shift_time VARCHAR(50)
    )
''')

print('Table created successfully!!')
```

```python
# Violation
cursor.execute('''CREATE TABLE violation (
    vehicle_number VARCHAR(20),
    violation_type VARCHAR(100),
    fine_amount FLOAT
    )
''')

print('Table created successfully!!')
```

```python
for _, row in df.iterrows():
    try:
        cursor.execute("""
            INSERT INTO police_logs 
            (stop_date, stop_time, country_name, driver_gender, driver_age, driver_race,
             search_conducted, search_type, drugs_related_stop, stop_duration,
             stop_outcome, violation, vehicle_number)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        """, (
            row['stop_date'],
            row['stop_time'],
            row['country_name'],
            row['driver_gender'],
            row['driver_age'],
            row['driver_race'],
            int(row['search_conducted']),
            row['search_type'],
            int(row['drugs_related_stop']),
            row['stop_duration'],
            row['stop_outcome'],
            row['violation'],
            row['vehicle_number']
        ))
    except Exception as e:
        print(f'Error inserting into police_logs: {e}')

connection.commit()  # Make sure to commit after inserts
```

```python
# Insert data

# Vehicles Data
for _ , row in df.iterrows():
    try:
        cursor.execute('''
            INSERT IGNORE INTO vehicles (vehicle_number, owner_id, type, flagged_status)
            VALUES(%s,%s,%s,%s)
            ''',(
                row['vehicle_number'] if 'vehicle_number' in row else f'VEH[row.name]',
                row.name,
                'private',
                bool(row['drugs_related_stop'])
            ))
    except Exception as e:
        print(f'Error inserting vehicles : {e}')
```

```python
# Officers Data

officers = [
    ('John Doe', 'Checkpoint A', 'Morning'),
    ('jane Smith', 'Checkpoint B', 'Night'),
    ('Mike Johnson', 'Checkpoint C', 'Evening')
]

for officer in officers:
    cursor.execute('''
        INSERT IGNORE INTO officers(name, post_location, shift_time)
        VALUES(%s,%s,%s)
        ''',officer)
```

```python
# Violation Data

for _,row in df.iterrows():
    try:
        cursor.execute('''
            INSERT INTO violation(vehicle_number, violation_type, fine_amount)
            VALUES(%s,%s,%s)
        ''', (
            row['vehicle_number'] if 'vehicle_number' in row else f'VEH[row.name]',
            row['violation'],
            500.00
        ))
    except Exception as e:
        print(f'Error inserting violation : {e}')
```

```python
# Check post logs Data

for _, row in df.iterrows():
    try:
        cursor.execute('''
            INSERT INTO check_post_logs(vehicle_number, driver_id, officer_id, timestamp, status)
            VALUES(%s,%s,%s,%s,%s)
            ''', (
            row['vehicle_number'] if 'vehicle_number' in row else f'VEH[row.name]',
            row.name,
            1,
            row['timestamp'],
            row['stop_outcome']
            ))
    except Exception as e:
        print(f'Error inserting Check post logs : {e}')
```

```python
connection.commit()
cursor.close()
connection.close()

print('Data inserted Successfully')
```

```python
cursor.execute('''CREATE VIEW traffic_stops AS
SELECT
    cpl.timestamp AS stop_time,
    DATE(cpl.timestamp) AS stop_date,
    v.post_location AS country_name,
    '' AS driver_gender, -- If you want to extend the schema
    NULL AS driver_age,
    '' AS driver_race,
    v.type AS violation_raw,
    v.type AS violation,
    FALSE AS search_conducted,
    NULL AS search_type,
    cpl.status AS stop_outcome,
    CASE 
        WHEN LOWER(cpl.status) = 'arrested' THEN TRUE
        ELSE FALSE
    END AS is_arrested,
    '6-15 min' AS stop_duration, -- Placeholder
    CASE 
        WHEN LOWER(v.flagged_status) = 'drugs' THEN TRUE
        ELSE FALSE
    END AS drugs_related_stop
FROM check_post_logs cpl
JOIN vehicles v ON cpl.vehicle_number = v.vehicle_number;
''')
```
